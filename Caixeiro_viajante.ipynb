```python
import random
```

```python
## Gerar uma população inicial
# Cada cromossomo será representado pela sequência de pontos a serem percorridos

def initial_population(size):
  #elementos = ''.join(str(e) for e in labels)
  elementos = np.arange(0,POINTS,1, dtype=int)

  caminhos = []
  for i in range(0,size):
    #caminhos.append(''.join(random.sample(elementos,len(elementos))))
    np.random.shuffle(elementos)
    caminhos.append([e for e in elementos])

  return np.array(caminhos)


caminhos = initial_population(size = 20)
caminhos
```

```python
caminhos[0,:]
```

```python
## Calcular o fitness de cada indíviduo da população
#  Fitness = a distância total a ser percorrida por cada caminho

def calc_dist(pontos):
  dist = 0

  for i in range(1, len(pontos)):
    dist += np.linalg.norm(x[int(pontos[i])] - x[int(pontos[i-1])])

  dist += np.linalg.norm(x[int(pontos[-1])] - x[int(pontos[0])])

  return dist


def fitness(caminhos, size, verbose=False):
  
  fitness = np.zeros((size,))
  for i in range(0,len(caminhos)):
    
    fitness[i] = calc_dist(caminhos[i,:])

  population = np.column_stack((caminhos, fitness))
  population = population[np.argsort(population[:,-1])]
  
  return population


population = fitness(caminhos, size=20)
population
```

```python
def torneio(population, k, size):

  p = np.random.choice(np.arange(0,size), size=k)

  c = np.array([population[p[i],:] for i in range(k)])

  c = c[c[:, -1].argsort()]

  return list(c[0,:POINTS])


def crossover(population, size, crossover_rate=0.6, k=3, elit=1):
  new_population = []

  for i in range(elit):
    new_population.append(list(population[i,:POINTS]))


  while(len(new_population) < size):
    p1 = torneio(population, k, size)
    p2 = torneio(population, k, size)


    if(np.random.random() < crossover_rate):
      crossoverpoint = random.randint(0,len(p1))
      crossoverpoint2 = random.randint(0,len(p1))
      f = []

      count = 0
      for i in p1:
        if(count == crossoverpoint):
          break
        if(i not in p2[crossoverpoint:crossoverpoint2]):
          f.append(i)
          count = count + 1
    
      f.extend(p2[crossoverpoint:crossoverpoint2])
            
      f.extend([x for x in p1 if x not in f])

      #f1 = p1[0:crossoverpoint] + p2[crossoverpoint:10]
      #f2 = p2[0:crossoverpoint] + p1[crossoverpoint:10]

      new_population.append(f)

      #new_population.append(f1)
      #new_population.append(f2)

    else:
      new_population.append(p1)
      new_population.append(p2)

    if(len(new_population) > size):
      del new_population[-1]


  return np.array(new_population)


crossover(population, size=10, crossover_rate=1)

```

```python
def mut(x):
  mutpoint1 = random.randint(0,POINTS-1)
  mutpoint2 = random.randint(0,POINTS-1)

  #print(x)

  aux = x[mutpoint1]
  x[mutpoint1] = x[mutpoint2]
  x[mutpoint2] = aux

  #print(x)

  return list(x)

def mutation(population, mutation_rate=0.1):
  new_cromossomos = []
  new_cromossomos.append(list(population[0]))

  for element in population[1:]:
    if(np.random.random()<mutation_rate):
      new_cromossomos.append(mut(element))
    else:
      new_cromossomos.append(element)

  cromossomos = []
  for element in new_cromossomos:
    cromossomos.append([int(i) for i in element])

  return np.array(cromossomos)


mutation(population[:,:POINTS], mutation_rate=1)
```
